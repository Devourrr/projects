# 엑셀 다루기
엑셀 구조 Application> Workbook(파일)> Sheeets> Range, Rows, Columns, Cells...

```python
from openpyxl import Workbook
'''
데이터 쓰기
'''

wb = Workbook() # 워크북 생성
ws = wb.active # 활성화된 시트 선택
ws.title = 'sample' #시트명 지정

# 셀 주소로 값 입력
ws['A1'] = 'KJh says'


# 셀 좌표로 값 입력(행, 열, 값)
ws.cell(1,2 ,"Slow and Steady Wins the race.")

import datetime
ws['A2'] = datetime.datetime.now()
ws['A3'] = datetime.datetime.today()

# 행 단위로 값 입력(마지막에 행 추가)
ws.append([1,2,3])
ws.append([4,5,6])
ws.append([7,8,9])

#수식 입력
ws.cell(4,4,'=sum(A4:C4)')
ws.cell(5,4,'=sum(A5:C5)')
ws.cell(6,4,'=sum(A6:C6)')


# 엑셀 파일 저장
wb.save('sample1.xlsx')
```

```python
'''
데이터 읽기
'''

from openpyxl import load_workbook
# wb = load_workbook("sample1.xlsx",data_only=True)# data_only=True: 수식은 제외하고 값만 불러오기
wb = load_workbook("sample1.xlsx",data_only=False) # 수식+값 불러오기
ws = wb['sample'] # 시트 이름으로 불러오기

# 셀 주소로 값 출력
print(ws["A1"].value)
print()
print(ws["A4"].value,ws["B4"].value,ws["C4"].value,ws["d4"].value)

# 셀 좌표로 값 출력(행, 열)
print()
print(ws.cell(1,2).value)

# 지정한 셀의 값 출력
print()
cells = ws["A4":"D6"]

print()
print(type(cells))
print()
print(cells)
print()

for row in cells:   # row = 행, 행 하나씩 꺼내기
    for cell in row: # cell = 열, 해당 행에서 열 하나씩 꺼내기
        print(cell.value) # 해당 셀 값 출력
print()       
# 모든 행 단위로 출력
print(ws.rows)

for row in ws.rows: # 행을 기준으로 출력
    print(row)
print()

for column in ws.columns: # 열을 기준으로 출력
    print(column)
print()

# 모든 행과 열 출력
all_values = []
for row in ws.rows:
    row_value = []
    for cell in row:
        row_value.append(cell.value)
    all_values.append(row_value)
print(all_values)
```

```python
import requests
from bs4 import BeautifulSoup
from openpyxl import Workbook

url = "https://unse.daily.co.kr/unse/?p=zodiac"
res = requests.get(url)
res.raise_for_status()
bs = BeautifulSoup(res.text,'lxml')

wb = Workbook()
ws = wb.active

for i in range(12):

    totem = bs.select_one(f'#code_{i+1} b').get_text()
    print(totem)
    ws.title = totem
                          
    ws['A1'] = "년도"
    ws['B1'] = "운세"
    ws['A2'] = "요약"
                          
    summary = bs.select(f"#code_{i+1} p").get_text()
    ws['B2'] = summary
                          
    spec = bs.select_one(f"#code_{i+1}>li")[1:]
        #list = {}
    for i in range(len(spec)):
        year = spec[j].select_one("span").get_text()
        text = spec[j].select_one("p").get_text()
                         #lsit[year] = text
        ws[f'A{j+3}'] = year
        ws[f'B{j+3}'] = text
                         
    if i <11:
        ws = wb.create_sheet()
                         
wb.save('오늘의운세.xlsx')
                          
    
```
