```python
import requests
from bs4 import BeautifulSoup
date = input('날짜입력(yyyymmdd):')
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20210423'
res = requests.get(url)

print(res)

# HTML 파서 선택
# bs = BeautifulSoup(res.text,'html.parser') # 속도빠름
bs = BeautifulSoup(res.text,'lxml')
# bs = BeautifulSoup(res.text,'html.5lib') # 표준규격 #select는 list 형태로 출력

print(f'{date}의 영화 평점 랭킹')

# 부모 요소 찾기
tbody = bs.select_one('#old_content > table > tbody')
trs = tbody.select('tr') # select 함수는 list형태로 출력
# print(trs)
# print(len(trs)) tr아래 자식td 개수 찾기, 불필요한 개수 추려내기

for tr in trs:
    if len(tr)<2: #tr 아래 자식 td가 2개 미만이면
        continue # 다음 tr로 넘어감
        
    rank=tr.select_one('img')['alt']
    name = tr.select_one('a').get_text()
    score =tr.select_one('td.point').get_text() # new 가장 아래 td -가 밑에서 두번째
    change = tr.select_one('td:nth-child(6) > img')['alt']
    
    if len(tr) == 17:
        change_range ='▣'
    else:
        change_range = 'Chg_rate:'+ tr.select_one('td.range.ac').get_text() #none 타입 에러
        
    if change =='na':
        change = change.replace('na','-')
    elif change =='up':
        change = change.replace('up','△')
    elif change =='down':
        change =change.replace('down','▽')
    elif change == 'new':
        change = change.replace('new','NEW')
    
    
    print(rank, name, score, change, change_range)
    
```
