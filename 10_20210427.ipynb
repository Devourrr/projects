```python
import requests
from bs4 import BeautifulSoup

url = "http://www.dietshin.com/calorie/sports_main.asp"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.85 Safari/537.36"}

res = requests.get(url, headers = headers)
print(res)
print('============================================================================')    

soup = BeautifulSoup(res.text, 'lxml')

charts = soup.select_one('#container > div.contents.calorieDc > table > tbody')
# print(charts)

ranks = charts.select('tr')
# print(len(ranks))

chart_lst = []
for row in ranks:
    columns = row.select('td')
#     print(len(columns))
    rank =  columns[0].get_text()+'.'
    rank_var = columns[1].select_one('img')['alt']
    if rank_var == "동등":
            rank_var = '-'
    elif rank_var == "상승":
            rank_var = "▲"
    elif rank_var == "하락":
            rank_var = "▼"    
        
    exercise = columns[2].get_text()
    how = columns[3].get_text()
    time = columns[4].get_text()
    kcal = columns[5].get_text()
    
    print(rank, rank_var, exercise, how, time, kcal)
    chart_lst.append([rank, rank_var, exercise, how, time, kcal])
    
print('============================================================================')    
    
# 데이터 출력
key = input("운동 단위를 입력하세요. (ex.회/분):  ")

print(f">>> 운동 단위(회/분): {key} \n")
print("<금주의 운동칼로리 검색 순위 TOP>")

i = 0
for row in chart_lst:
    if key in row[4]:
        i += 1
        print(str(i)+'.', (' ').join(row[1:]))
print('============================================================================')
```
